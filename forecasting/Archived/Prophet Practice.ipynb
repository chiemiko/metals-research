```python
import pandas as pd
from fbprophet import Prophet
import numpy as  np

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
sales_df = pd.read_csv('forecasting_raw_data/example_retail_sales.csv')
```

```python
sales_df.head()
```

```python

sales_df['y_orig'] = sales_df['y'] # to save a copy of the original data..you'll see why shortly. 
# log-transform y
sales_df['y'] = np.log(sales_df['y'])
sales_df.head()
```

```python
model = Prophet() #instantiate Prophet
model.fit(sales_df); #fit the model with your dataframe
```

```python
future_data = model.make_future_dataframe(periods=10, freq = 'm')
```
