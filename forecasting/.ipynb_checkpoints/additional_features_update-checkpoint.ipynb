```python
# Example r1_list values are: 
# ['000920', '00047', '000', '000110', '000', '000']
def parse_for_world_consumption(file):
    '''Input: filename, Output: list of four digits of World Values'''
    try:

        pdfFileObj = open(file, 'rb') 
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

        pageObj = pdfReader.getPage(1) 
        string = pageObj.extractText()
        r1 = re.findall(r"\(rounded\).+", string, re.MULTILINE)
        r1 = r1[0]
        r1_list = list(r1.split(','))
        r1_list


        first_num = re.findall(r"\d*", r1_list[0])
        # Note: no need for last digit in this stripping because last section doesn't have letters at end
        #last_num = re.findall(r"\d*", r1_list[len(r1_list)-1])
        first_num


        '''Strip split'''
        first_digit = [int(s) for s in first_num if s.isdigit()]
        first_digit = int(first_digit[0])
        #print(r1_list[1:])
        #print(first_digit)

        china_values = []

        #print(r1_list)
        #print()
        for item in r1_list[1:]:
            if len(item) == 3:
                '''If number of values is 3, append to last set of numbers'''
                added_string_result = str(first_digit)+ item
                first_digit= added_string_result
                result = int(added_string_result)
                # Delete previous repeated values in the results list
                try:
                    if added_string_result[:len(str(china_values[-1]))] == str(china_values[-1])[:len(str(china_values[-1]))]:
                        china_values = china_values[:-1]

                except IndexError:
                    pass

                china_values.append(result)

            else:
                '''If number in list is more than 3 digits, split up into first 3 values and append that, 
                then separate to a new number'''
                added_string_result = str(first_digit) + item[0:3]
                # Delete previous repeated values in the results list
                try:
                    if added_string_result[:len(str(china_values[-1]))] == str(china_values[-1])[:len(str(china_values[-1]))]:
                        china_values = china_values[:-1]

                except IndexError:
                    pass
                result = int(added_string_result)
                china_values.append(result)
                first_digit = item[3:]
                #print(first_digit)

        return china_values


    except TypeError:
        print("ERROR from file: ", file)
        print()
        pass
        
```
